```python
# imports and packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# Load dataset
df = pd.read_csv(r"dataset/transformer-voltage.csv",
                 index_col= 'Sample',
                 delimiter=',')
df
```

```python
# summary of dataframe
df.info()
```

```python
# summary of statistics
df.describe()
```

```python
# z-critical for the two-tailed test
alpha = 0.01
z_critical = stats.norm.ppf(1 - alpha/2)
z_critical
```

```python
# confidence interval
pop_std = 5

ci_lower, ci_upper = stats.norm.interval(
    confidence=0.95,
    loc = df['Voltage'].mean(),
    scale = pop_std/np.sqrt(df['Voltage'].count())
)

print(f"Confidence Interval: [{ci_lower:.2f}, {ci_upper:.2f}]")
```

```python
# voltage distribution
mean = df['Voltage'].mean()
std = df['Voltage'].std(ddof=1)

# x-axis
x_min = df['Voltage'].min()
x_max = df['Voltage'].max()

# normal distribution plot
x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)
plt.plot(x, p,
         linewidth=2,
         color="#A30000")

# histogram 
plt.hist(df['Voltage'],
         bins=5,
         density=True, 
         alpha=0.8,
         color="#0104C0")

plt.title('Voltage Distribution')
plt.ylabel('Probability Density')
plt.xlabel('Voltage(V)')

#confidence interval
plt.axvline(x=ci_lower,
            color='black', 
            linestyle='--')
plt.axvline(x=ci_upper,
            color='black', 
            linestyle='--')
plt.show()
```
